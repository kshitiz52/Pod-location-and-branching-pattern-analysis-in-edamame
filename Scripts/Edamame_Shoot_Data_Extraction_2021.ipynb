```python
pod_annotations = pod_df['annotations'] #annotation information
pod_single = json.loads(pod_annotations[776])[0]# annotation information of a single (last) image
#print(pod_single)
value_single = pod_single['value']# values of annotation information of a single (last) image, excludes the header/coumnn names
print(value_single)
```

```python
def distance(point1,point2):
  x1,y1 = point1
  x2,y2 = point2 
  return ((x1-x2)**2 + (y1-y2)**2 )**0.5

#parse from dict and get top, bottom, branching points and ruler as a tuple
def parseData(value):
  top = ()
  bottom = ()
  branching_points = []
  ruler=()
  for args in value: 
    if args['tool_label'] == 'Top':
      top = (args['x'],args['y'])#cartesian values of top of the plant
    if args['tool_label'] == 'Bottom':
      bottom = (args['x'],args['y'])#cartesian values of bottom of the plant
    if args['tool_label'] == 'Branching point':
      branching_points.append( (args['x'],args['y']))# There are many branching points so the information is in the form of list 
    if args['tool_label'] == 'Horizontal rule':
      ruler = ((args['x1'],args['y1']), (args['x2'],args['y2'])) #cartesian values of the two extreme points of the ruler
    
  return (top,bottom,branching_points,ruler)
```

```python
def calculateHeight(single_data, ruler_length=91.5):
  top,bottom,branching_points,ruler = parseData(single_data)#merge all data into a single_data
  length_cm = 91.5
  if len(ruler) != 2 or len(bottom)!=2 or len(top)!=2: #Return -1 if the data points of the ruler, top, bottom are not exactly equal to 2
    return (-1,-1,-1)

  cm_to_pixel_length = 91.5 / distance(ruler[0],ruler[1])# to convert pixels of ruler length to cm

  branching_points_with_dist = branching_points.copy()
  #arrange branches in order from bottom
  for i in range(len(branching_points)):
    branch = branching_points[i]
    if len(branch)!= 2:
      return (-1,-1,-1)
    branching_points_with_dist[i] = (distance(bottom, branch)*cm_to_pixel_length,branch)# Calculates the length of each branchng points from the bottom of the plant

  branching_points_with_dist.sort()

  plant_height_in_pixel = distance(top,bottom)# plant height or the distance between top and bottom of the plant in pixel
  plant_height_in_cm = cm_to_pixel_length * plant_height_in_pixel #plant height in cm

  # print(branching_points_with_dist) #length in cm
  # print(plant_height_in_cm) #height in cm
  branch_count = len(branching_points)#Total number of branches
  
  if branch_count == 0:# for the unbranchd plants
    first_branch_length = 0
  else:
    first_branch_length = branching_points_with_dist[0][0]# length of first branch from the bottom of the plant which is the First Internode length
  
  return (plant_height_in_cm, first_branch_length, branch_count)
  
#print("For this data point")
#print(value_single)
print("calculated Values for this data point viz. (Plant_Ht, FIL, # of Branches)")
print(calculateHeight(value_single))
```

```python
def heightCalculator(single_row,type='all'):
  annotation = single_row['annotations']
  single_data = json.loads(annotation)[0]
  # print(single_data)
  if not single_data['task'] == 'T1':
    output = (-1,-1,-1)
  else:
    output =  calculateHeight(single_data['value'])
  if type=='plant_height':
    return output[0]
  if type == 'first_branch_height':
    return output[1]
  if type=='branch_count':
    return output[2]
  return output
```

```python
def parseName(single_row):
  subject = json.loads(single_row['subject_data'])
  return subject[list(subject.keys())[0]]['Filename']
```

```python
output = pod_df.copy() 
output['Filename'] = output.apply(parseName,axis=1)
output['plant_height'] = output.apply(heightCalculator,axis=1,args=('plant_height',))
output['first_branch_height'] = output.apply(heightCalculator,axis=1,args=('first_branch_height',))
output['branch_count'] = output.apply(heightCalculator,axis=1,args=('branch_count',))

print(output) #new dataframe with an extra column of heights
output.to_csv("../Outputs/Edamame_shoot_architecture_output_data_2021.csv")
```
